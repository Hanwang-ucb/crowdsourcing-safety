## Streetstory

```python
streetstory = pd.read_csv('Street Story reports - 03_03_2021.csv')
streetstory.head(5)
```

```python
unsafe_text = streetstory['unsafe location narrative'].dropna()
unsafe_text
```

```python
from string import punctuation
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize.treebank import TreebankWordDetokenizer

stop_words = set(stopwords.words('english'))

porter = PorterStemmer()

def remove_punctuation(document):

    no_punct = ''.join([character for character in document if character not in punctuation])
    
    return no_punct

def remove_digit(document): 
    
    no_digit = ''.join([character for character in document if not character.isdigit()])
              
    return no_digit

def remove_stopwords(document, stop_words):
    removed = 0
    while removed == 0:
        removed = 1
        for stop_word in stop_words:
            try:
                document.remove(stop_word)
            except ValueError:
                pass
            else:
                removed = 0
    return document

def stemmer(document):
    stemmed_document = [porter.stem(word) for word in document]
    return stemmed_document
```

### Lowercase, remove punctuation, remove digits

```python
unsafe_text = unsafe_text.str.lower().apply(remove_punctuation).apply(remove_digit)
```

### Tokenize

```python
unsafe_text = unsafe_text.apply(word_tokenize)
unsafe_text.head(5)
```

### Remove stop words

```python
unsafe_text = unsafe_text.apply(lambda x: remove_stopwords(x, stop_words))
unsafe_text.head(5)
```

### Stemming

```python
unsafe_text = unsafe_text.apply(stemmer)
unsafe_text.head(5)
```

### De-tokenize

```python
unsafe_text = unsafe_text.apply(TreebankWordDetokenizer().detokenize)
unsafe_text.head()
```

### DTM

```python
from sklearn.feature_extraction.text import CountVectorizer

countvec = CountVectorizer(min_df=0.005)

sparse_dtm = countvec.fit_transform(unsafe_text)
sparse_dtm
```

```python
dtm = pd.DataFrame.sparse.from_spmatrix(sparse_dtm, columns=countvec.get_feature_names(), index=unsafe_text.index)
dtm
```

```python
dtm.sum().sort_values(ascending=False) 
```
